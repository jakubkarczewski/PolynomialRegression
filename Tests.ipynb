```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
my_data = np.genfromtxt('./ai-task-samples.csv', delimiter=',')
    
data_x_raw = np.zeros((len(my_data), 1))
data_y = np.zeros((len(my_data), 1))

for i in range(len(my_data)):
    data_x_raw[i] = my_data[i][0]
    data_y[i] = my_data[i][1]
    
plt.ion()

fig, ax = plt.subplots(1, 1)

ax.scatter(data_x_raw, data_y)
```

```python
scaling = [1.00000000e+00, 9.99144979e+01, 9.99921447e+03, 9.97437129e+05,
 9.99842900e+07, 9.95732199e+09]
weights = [ 2.76056803e+10,  6.86021250e+08, -3.91640658e+05,  6.62653362e+04,
  3.58345837e+02, -6.78878532e+00]
new_weights = [a/b for a,b in zip(scaling,weights)]
y_model = np.polyval(new_weights, np.linspace(-100,100,10000))
plt.plot(np.linspace(-100,100,10000), y_model, c='r', label='Model')
plt.scatter(data_x_raw, data_y, c='b', label='Training data')
polyfited = np.polyfit(data_x_raw.flatten(), data_y, 6)
polyval = np.polyval(polyfited, data_x_raw)
plt.scatter(data_x_raw, polyval, c='g', label='Polyvaled data')
plt.grid()
plt.legend(loc='best')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
